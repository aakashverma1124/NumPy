```python
# NumPy array 

import numpy as np

a = np.array([1,2,3])
print(a)
print(a.max()) # gives max of array
print(a.min()) # gives min of array
print(a.sum()) # gives sum of array
```

```python
# NumPy two dimensional array 

import numpy as np

a = np.array([(1,2,3),(5,6,7),(8,9,10),(2,3,5)])
print(a)

print(a.ndim) # gives dimension of array
print(a.itemsize) #gives size of each item in array
print(a.dtype) # gives datatype of array
print(a.size) # gives no. of elements in array
print(a.shape) # gives row,col of array
print(a.reshape(2,6)) # reshaped array in 2 rows and 6 columns, it doesn't alter the original array

# slicing operation in array

print(a[0,2]) # gives value of 0th row and 2nd column
print(a[0:,2]) # gives all 2nd column from all rows
print(a[0:2,2]) # gives rows from 0 to 1 and print 2nd column 

```

```python
# linespacing

a = np.linspace(1,50,10) # gives 10 equally spaced values from 1 to 50
print(a)
```

```python
a = np.array([(1,2,3),(4,5,6)])
print(a.sum(axis = 0)) # gives sum of columns
print(a.sum(axis = 1)) # gives sum of rows
```

```python
# square root, standard deviation and ravel operation

a = np.array([(4,9,16),(25,36,49)])
print(np.sqrt(a)) # gives square root of all elements of array

print(np.std(a)) # gives standard deviation 

print(a.ravel()) # converts whole array into 1 column
```

```python
# arithmetic operation in NumPy

a = np.array([(1,2,3),(4,5,6)])
b = np.array([(1,2,3),(4,5,6)])

print(a + b) # Remember! In list, this operation concatenates two list.
print(a - b)
print(a * b)
print(a / b)


# stacking in NumPy array

print(np.hstack((a,b))) # horizontal stacking
print(np.vstack((a,b))) # vertical stacking
```

```python
# working with speacial functions........

import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y = np.sin(x)


plt.plot(x,y)
plt.show()

y = np.cos(x)

plt.plot(x,y)
plt.show()

y = np.exp(x)

plt.plot(x,y)
plt.show()
```

```python
# this code shows that numPy array takes less memory as compared to list.
import numpy as np
import time
import sys

ls = range(1000)
print(sys.getsizeof(5)*len(ls))

ds = np.arange(1000)
print(ds.size*ds.itemsize)
```

```python
# this example shows that : 
# 1. numPy array is faster than list.
# 2. it is more convinient to work with numPy array than list. 

import numpy as np
import time
import sys

size = 10000000

L1 = range(size)
L2 = range(size)

A1 = np.arange(size)
A2 = np.arange(size)

start = time.time()
result = [(x + y) for x, y in zip(L1, L2)]  # writing code for adding two lists is not much convinient as compared to adding two numPy array .
print((time.time() - start)*1000)

start = time.time()
result = A1 + A2  # easy to work with numPy array
print((time.time() - start)*1000) 

# by multiplying with 1000, the time is converted from seconds to milliseconds.
```
